```python
dataset.drop(columns=["Cabin"],inplace=True)
```

```python
dataset.drop(columns=["PassengerId"],inplace=True)
```

```python
dataset.head(3)
```

```python
#qki agr mainn dropna use krti hu too puri row delete ho skti h to apn fillna ka use krengee
```

```python
dataset["Age"]=dataset["Age"].fillna(dataset["Age"].mode()[0],inplace=True)
dataset
```

```python
dataset["Embarked"]=dataset["Embarked"].fillna(dataset["Embarked"].mode()[0],inplace=True)
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.drop(columns=["Name"],inplace=True)
```

```python
dataset.drop(columns=["Ticket"],inplace=True)
```

```python
dataset.head(3)
```

```python
#we use map function for encoding
dataset["Sex"]=dataset["Sex"].map({"male":0,"female":1})
```

```python
dataset.info()
```

```python
dataset["Embarked"].value_counts()
```

```python
dataset["Embarked"]=dataset["Embarked"].map({"S":0,"C":1,"Q":2})
```

```python
dataset.info()
```

```python
dataset.head(10)
```

```python
#finding the outliers using boxplot
sns.boxplot(x="Fare",data=dataset)
plt.show()
```

```python
q1=dataset["Fare"].quantile(0.25)
q1
```

```python
q3=dataset["Fare"].quantile(0.75)
q3
```

```python
iqr=q3-q1
```

```python
min_range=q1-(1.5*iqr)
max_range=q3+(1.5*iqr)
min_range,max_range
             
```

```python
new_dataset=dataset[(dataset["Fare"]>=min_range) & (dataset["Fare"]<=max_range)]
```

```python
dataset.shape
```

```python
new_dataset.shape
```

```python
sns.boxplot(x="Fare",data=new_dataset)
plt.show()
```

```python
#removing duplicated values from new_dataset....first check is their exist any duplicate valuess
new_dataset.duplicated().sum()
```

```python
new_dataset=new_dataset.drop_duplicates()
```

```python
new_dataset.duplicated().sum()
```

```python
#we dont have any duplicate valuess.....
```

```python
#now we split the data can be two ways slicing(using iloc) or drop columns but we use drop columns becoz we donthave survived column in the endd and i dont want to do that so called messy work by manually
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression


x=new_dataset.drop(columns=["Survived"])
y=new_dataset["Survived"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_test
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
scaler=StandardScaler()#feature scaling
```

```python
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train_scaled,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred=model.predict(x_test_scaled)
```

```python
y_pred
```

```python
predictions=pd.Series(y_pred)
```

```python
predictions.value_counts()
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
