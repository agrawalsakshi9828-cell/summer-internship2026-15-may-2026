```python
from sklearn.model_selection import train_test_split
x = df[['Store',  'Holiday_Flag', 'Temperature', 'Fuel_Price',
       'CPI', 'Unemployment', 'year', 'month', 'date']]
y = df["Weekly_Sales"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.20, random_state= 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor( n_estimators=500,max_depth=20,min_samples_split=5,random_state=42)
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)*100, model.score(x_test, y_test)*100
```

```python
from sklearn.metrics import r2_score
y_pred = model.predict(x_test)
r2 = r2_score(y_test, y_pred)
print("R² Score:", r2)
```
