# cropping image

```python
image=cv2.imread("disney.png")
```

```python
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
start_row, end_row=50, 200
```

```python
start_col, end_col=100, 300
```

```python
cropped=image_rgb[start_row:start_col,end_row:end_col]
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,4))
axes[0].imshow(image_rgb)
axes[0].set_title("original image")
axes[0].axis("off")

axes[1].imshow(cropped)
axes[1].set_title("cropped image")
axes[1].axis("off")

plt.show()
```

# resize image and rotation the image

```python
image=cv2.imread("disney.png")
```

```python
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
height,width=image.shape[:2]
```

```python
center=(width//2,height//2)
```

```python
rotation_image=cv2.getRotationMatrix2D(center,45,1)
```

```python
rotated=cv2.warpAffine(image,rotation_image,(width,height))
```

```python
rotated_rgb=cv2.cvtColor(rotated,cv2.COLOR_BGR2RGB)
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,4))
axes[0].imshow(image_rgb)
axes[0].set_title("original image")
axes[0].axis("off")

axes[1].imshow(rotated_rgb)
axes[1].set_title("rotated image")
axes[1].axis("off")

plt.show()

```

```python
import cv2
import numpy as np

canvas = np.ones((400, 600, 3), dtype="uint8") * 255

# Saffron strip
cv2.rectangle(canvas, (150, 100), (450, 150), (0, 128, 255), -1)

# White strip
cv2.rectangle(canvas, (150, 150), (450, 200), (255, 255, 255), -1)

# Green strip
cv2.rectangle(canvas, (150, 200), (450, 250), (0, 255, 0), -1)

# Ashoka Chakra
cv2.circle(canvas, (300, 175), 20, (255, 0, 0), 2)

# Flag pole
cv2.line(canvas, (150, 80), (150, 320), (0, 0, 0), 3)

cv2.imshow("Indian Flag", canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
