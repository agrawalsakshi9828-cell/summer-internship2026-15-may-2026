```python
import random
computer_choice=random.randint(1,10)
while True:
    print("""
    1.to try luck
    2.bhai drr lgg rha h
    """)
    choice=int(input("bata bhai kya krna h: "))
    if choice==1:
        user_value=int(input("enter number from 1 to 10: "))
        if user_value<=10:
            if computer_choice==user_value:
                print("congratulation you win")
            else:
                print("better luck next time")
        else:
            print("enter correct value 1 to 10")
    elif choice==2:
        print("bhai tu drr gaya")
        break
    else:
        print("shi option select krr le")
```
