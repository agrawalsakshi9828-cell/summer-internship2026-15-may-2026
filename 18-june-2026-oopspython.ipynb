# multilevel inheritance

```python
#multilevel inheritance
class grandfather:
    def land(self):
        print("20 arcs of lands")
        
```

```python
class father(grandfather):
    def house(self):
        print("VILLA")
```

```python
class child(father):
    def car(self):
        print("BMW")
```

```python
s1=child()
```

```python
s1.car()
s1.house()
s1.land()
```

# super keyword

```python
#super keyword value ko override nhi krna parent wali kaaa
class parent:
    def display(self):
        print("i am parent class")
```

```python
class child(parent):
    def display(self):
        super().display()
        print("i m child class")
```

```python
ch=child()
```

```python
ch.display()
```

# constructor and super keyword

```python
#constructor and super keyword
class parent:
    def __init__(self):
        print("parent class created")
class child(parent):
    def __init__(self):
        #parent wale constructor ko child class mein call krna h 
        super().__init__()
        print("child class")
```

```python
c1=child()
```

# getter method

```python
#getter and setter method
class bank:
    def __init__(self):
        self.balance=500
    def get_balance(self):
        return self.__balance
```

```python
b1=bank()
```

```python
b1.get_balance#simply hiding the value or balannce using getter method
```

# setter method

```python
#setter method
class bank:
    def __init__(self):
        self.balance=500
    def deposit(self,amount):
        self.balance += amount
    def get_balance(self):
        return self.balance
```

```python
b1=bank()
```

```python
b1.deposit(300)
```

```python
b1.get_balance
```
