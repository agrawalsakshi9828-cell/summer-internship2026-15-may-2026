```python
Y=df[["Class"]]
```

```python
x=df.drop("Class",axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
Y=label_encoder.fit_transform(Y)
Y
```

```python
label_encoder.classes_
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, Y_train, Y_test = train_test_split(
    x,
    Y,
    test_size=0.20,
    random_state=42,
    stratify=Y
)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(
    Dense(64, activation='relu', input_shape=(16,))
)
```

```python
model.add(
    Dense(32,activation="relu")
    
)
```

```python
model.add(
    Dense(16,activation="relu")
    
)
```

```python
model.add(
    Dense(8,activation="relu")
    
)
```

```python
model.add(
    Dense(7,activation="softmax")
    
)
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
train_model=model.fit(x_train,Y_train,epochs=50, #repeation
    batch_size=32,
    validation_split=0.2,)
```

```python
loss,accuracy=model.evaluate(x_test,Y_test)
```

```python
print(loss)
```

```python
print(accuracy*100)
```

```python
plt.figure(figsize=(10,5))

plt.plot(train_model.history["accuracy"])
plt.plot(train_model.history["val_accuracy"])

plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["Train", "Validation"])

plt.show()
```

```python
predictions=model.predict(x_test)
```

```python
prediction_class=predictions.argmax(axis=1)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(Y_test,prediction_class))
```

```python
cm=confusion_matrix(Y_test,prediction_class)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(cm,annot=True,fmt="d",cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
