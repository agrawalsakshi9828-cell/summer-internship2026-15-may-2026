```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df = pd.DataFrame({"name" : ["a", "b", "c", "d"]})
df
```

```python
le.fit(df)
```

```python
le.transform(df)
```

```python
df["e_name"] = le.fit_transform(df)
```

```python
df.drop(columns = "name", inplace = True)
```
