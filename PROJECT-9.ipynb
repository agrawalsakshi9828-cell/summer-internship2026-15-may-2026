```python
def predict_comment (Comment):
    cmt = Comment.lower()
    cmt = re.sub(r'[^a-z \s]', " ", cmt)
    token = word_tokenize(cmt)
    token = [w for w in token
            if w not in stopwords]
    tokens = [stemmer.stem(w) for w in token]
    clean_cmt = " ".join(tokens)
    vec = tfidf.transform([clean_cmt])
    result = model.predict(vec)[0]
    prob= model.predict(vec)[0]
    return result, prob
```
