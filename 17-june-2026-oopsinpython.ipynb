```python
class student:
    def add_data(self,name,age):
        self.name=name
        self.age=age
    def display(self):
        print("name",self.name)
        print("age",self.age)
    def voting(self):
        if self.age>=18:
            print(self.name,"can vote")
        else:
            print(self.name,"cant vote")
```

```python
rahul=student()
```

```python
rahul.add_data("rahul",23)
```

```python
rahul.display()
```

```python
rahul.voting()
```

```python
class bankaccount:
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance
    def deposit(self,amount):
        self.balance=self.balance+amount
    def withdraw(self,amount):
        self.balance=self.balance-amount
    def show(self):
        print("name",self.name)
        print("balance",self.balance)   
```

```python
rahul=bankaccount("rahul",10000)
```

```python
rahul.show()
```

```python
rahul.deposit(1000)
```

```python
rahul.withdraw(500)
```

```python
rahul.show()
```
