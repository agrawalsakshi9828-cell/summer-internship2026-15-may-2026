```python
#finding the particular word is existing in file or not
word="heyyyy"
with open("k.txt","r") as f:
    data=f.read()
    if(data.find(word)!= -1):
        print("found")
    else:
        print("not found")
    print(data)
```

```python
def create_new_password():
    pw=input("enter the new password:")
    with open("password.txt","w") as file:
        file.write(pw)
    print("password created successfully")

def create_secrte_msg():
    user_pw=input("enter the pwd to write in file:")
    with open("password.txt","r") as file:
        saved_pwd=file.read()
        if user_pw==saved_pwd:
            secret_text=input("enter yoyr secret note:")
            with open("data.txt","w") as file:
                file.write(secret_text)
        else:
            print("incorrect pwd")

def read_secrte_msg():
    user_pw=input("enter the pwd to read in file:")
    with open("password.txt","r") as file:
        saved_pwd=file.read()
        if user_pw==saved_pwd:
            with open("data.txt","r") as file:
                print(file.read())
        else:
            print("incorrect pwd")

while True:
    print("""
1.create a pwd
2.write your secret note
3.read your secret note
4.exit
""")
    choice=input("tell me what to do:")
    if choice=="1":
        create_new_password()
    elif choice=="2":
        create_secrte_msg()
    elif choice=="3":
        read_secrte_msg()
    elif choice=="4":
        print("you exit")
        break
    else:
        print("bhai dhng se press krr le number")
        
```
